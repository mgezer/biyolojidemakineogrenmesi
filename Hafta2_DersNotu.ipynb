### **Büyük Sigma (Σ) ve Büyük Pi (Π) Gösterimi**

Bu gösterimler, matematiksel işlemlerde sıkça kullanılan sembollerdir. Özellikle makine öğrenmesi, istatistik ve biyolojik verilerin analizinde toplama ve çarpma işlemleri için kullanılırlar. Aşağıda, dökümanınızdaki bilgilerden yararlanarak bu iki gösterimi detaylı bir şekilde açıklayacağım.

---

### **1. Büyük Sigma (Σ) Gösterimi**
**Büyük Sigma**, bir koleksiyonun veya vektörün elemanlarının toplamını ifade etmek için kullanılır.

#### **Tanım:**
Sigma gösterimi şu şekilde tanımlanır:
$$
\sum_{i=1}^{n} x_i \stackrel{\text{def}}{=} x_1 + x_2 + \dots + x_n
$$
- **Σ:** Yunan alfabesinin "S" harfi olan sigma, toplamı temsil eder.
- **i=1:** Toplama işleminin başlangıç indeksi.
- **n:** Toplama işleminin bitiş indeksi.
- **xᵢ:** Toplanacak terimlerin genel ifadesi.

### Makine Öğrenmesindeki Anlamı:

-   **Toplam Hata Hesaplama**: Bir modelin tahminlerindeki hataların toplamı.
-   **Gradient Hesaplama**: Gradyan inişi (gradient descent) algoritmasında parametre güncellemeleri.
-   **Olasılık Hesaplama**: Olasılık dağılımlarının normalizasyonu.

### Örnek:


### **Lineer Regresyonda Hata Toplamı:**
Lineer regresyonda, modelin tahminleri ile gerçek değerler arasındaki farkların karelerinin toplamı (MSE - Mean Squared Error) şu şekilde hesaplanır:

$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

- **$y_{i} :$** Gerçek değer (hedef değişken).
- **$\hat{y}_i$:** Modelin tahmin ettiği değer.
- **$n$:** Veri setindeki örnek sayısı.

Bu formülde, her bir örnek için gerçek değer$y_{i} :$ile tahmin edilen değer $\hat{y}_i$ arasındaki farkın karesi alınır. Daha sonra bu karelerin toplamı hesaplanır ve örnek sayısına bölünerek ortalama hata bulunur.

---

#### **Örnek Uygulama:**
Diyelim ki aşağıdaki verilerimiz var:
- Gerçek değerler: $y = [3, 5, 7]$
- Tahmin edilen değerler: $\hat{y}_i= [2.5, 4.8, 6.9]$

Bu durumda MSE'yi hesaplayalım:

1. Her bir örnek için farkların karesini hesaplayın:


   $(y_1 - \hat{y}_1)^2 = (3 - 2.5)^2 = 0.25$

   $(y_2 - \hat{y}_2)^2 = (5 - 4.8)^2 = 0.04$

   $(y_3 - \hat{y}_3)^2 = (7 - 6.9)^2 = 0.01$

3. Karelerin toplamını alın:
  $\text{Toplam Kareler} = 0.25 + 0.04 + 0.01 = 0.30$
4. Örnek sayısına bölerek MSE'yi bulun:
   $\text{MSE} = \frac{0.30}{3} = 0.10$

````python
import numpy as np

# Gerçek değerler ve tahmin edilen değerler
y_true = np.array([3, 5, 7])
y_pred = np.array([2.5, 4.8, 6.9])

# MSE hesaplama
mse = np.mean((y_true - y_pred) ** 2)
print("Mean Squared Error (MSE):", mse)  # Çıktı: 0.10
````

### **Makine Öğrenmesindeki Anlamı:**
- **Toplam Hata Hesaplama:** Yukarıdaki örnekte gösterildiği gibi, MSE modelin performansını değerlendirmek için kullanılır. Düşük MSE değeri, modelin daha iyi performans gösterdiğini belirtir.
- **Gradient Hesaplama:** Gradyan inişi algoritmasında, parametrelerin güncellenmesi için hata fonksiyonunun türevi alınır. Bu süreçte büyük sigma gösterimi, toplam hatanın hesaplanmasında kullanılır.
- **Olasılık Hesaplama:** Büyük sigma gösterimi, olasılık dağılımlarının normalizasyonunda da kullanılır. Örneğin, bir olasılık kütle fonksiyonunun (pmf) toplamının 1'e eşit olması gerekir:
  $$
  \sum_{i=1}^{k} P(X = x_i) = 1
  $$


## **1. Büyük Pi (Π) Gösterimi**

### **Tanım**
**Büyük Pi (Π)** gösterimi, bir dizi sayının çarpımını ifade etmek için kullanılır. Bu gösterim, özellikle olasılık hesaplamaları ve makine öğrenmesindeki bazı modellerde sıklıkla karşımıza çıkar.

#### **Matematiksel Gösterim:**
$$
\prod_{i=1}^{n} x_i = x_1 \cdot x_2 \cdot \dots \cdot x_n
$$

- **$Π:$** Yunan alfabesinin "P" harfi olan pi, çarpımı temsil eder.
- **$i=1:$** Çarpma işleminin başlangıç indeksi.
- **$n:$** Çarpma işleminin bitiş indeksi.
- **$x_i:$** Çarpılacak terimlerin genel ifadesi.

Bu gösterimde, \( x_i \) terimleri belirtilen aralıkta (\( i=1 \)’den \( i=n \)'e kadar) çarpılır.

---

### **Makine Öğrenmesindeki Anlamı**
Büyük Pi gösterimi, makine öğrenmesinde özellikle olasılık hesaplamalarında ve model uygunluğunu değerlendirmede kullanılır:

1. **Olasılık Hesaplama:**
   Birden fazla bağımsız olayın birlikte gerçekleşme olasılığını hesaplamak için kullanılır. Örneğin:
   $$
   P(A \cap B \cap C) = \prod_{i=1}^{n} P(A_i)
   $$

2. **Likelihood (Olabilirlik):**
   Bir modelin verilere ne kadar iyi uyduğunu ifade eden olasılık hesaplamalarında kullanılır. Özellikle Naive Bayes sınıflandırıcılar bu gösterimi yoğun olarak kullanır.

3. **Naive Bayes Sınıflandırıcı:**
   Naive Bayes'te, bir sınıfa ait olma olasılığı, özelliklerin koşullu olasılıklarının çarpımı ile hesaplanır:
   $$
   P(y \mid x_1, x_2, \dots, x_n) \propto P(y) \prod_{i=1}^{n} P(x_i \mid y)
   $$
   - **$P(y)$:** Sınıfın önsel olasılığı.
   - **$P(x_i \mid y)$:** Her bir özelliğin sınıfa göre koşullu olasılığı.

---

### **Örnek: Naive Bayes Sınıflandırıcı**
Spam ve spam olmayan e-postalarda belirli kelimelerin görülme olasılıklarını hesaplayalım:


````python
# Spam ve spam olmayan e-postalarda kelime olasılıkları
P_spam = 0.3  # Spam olasılığı
P_not_spam = 0.7  # Spam olmama olasılığı

# Kelimelerin spam ve spam olmayan e-postalarda görülme olasılıkları
P_word1_spam = 0.8
P_word2_spam = 0.6
P_word1_not_spam = 0.1
P_word2_not_spam = 0.3

# Spam olasılığı
P_spam_given_words = P_spam * P_word1_spam * P_word2_spam

# Spam olmama olasılığı
P_not_spam_given_words = P_not_spam * P_word1_not_spam * P_word2_not_spam

print(f"Spam Olasılığı: {P_spam_given_words}")
print(f"Spam Olmama Olasılığı: {P_not_spam_given_words}")
````

## **2. Büyük Sigma (Σ) ve Büyük Pi (Π) Gösterimlerinin Birlikte Kullanımı**

### **Log-Likelihood Hesaplama**
Makine öğrenmesinde, özellikle model eğitimi sırasında log-likelihood hesaplamaları yapılır. Logaritma, çarpım işlemlerini toplama işlemlerine dönüştürerek hesaplamaları kolaylaştırır.

#### **Formül:**
$$
\text{Log-Likelihood} = \sum_{i=1}^{n} \log \left( \prod_{j=1}^{m} P(x_j \mid y_i) \right)
$$

- **$P(x_j \mid y_i)$:** Her bir özelliğin sınıfa göre koşullu olasılığı.
- **$\log $:** Çarpım işlemini toplama işlemine dönüştürmek için kullanılır.

---

### **Python Örnek: Log-Likelihood Hesaplama**
Aşağıdaki kod, bir modelin log-likelihood değerini hesaplar:

````python
import numpy as np

# Özelliklerin koşullu olasılıkları
P_features_given_class = np.array([[0.8, 0.6], [0.1, 0.3]])

# Log-likelihood hesaplama
log_likelihood = np.sum(np.log(np.prod(P_features_given_class, axis=1)))

print(f"Log-Likelihood: {log_likelihood}")
````

## **1. Vektör Nedir?**
Bir **vektör**, sıralı bir skaler değerler listesidir. Örneğin:
- Bir vektör $\mathbf{x} = [x^{(1)}, x^{(2)}, \dots, x^{(m)}]$ şeklinde gösterilir.
- $x^{(j)}$, vektörün $j$-inci boyutundaki değeri temsil eder.

**Biyolojiden Örnek:**
- Bir hücredeki 5 farklı genin ifade seviyelerini temsil eden bir vektör şu şekilde olabilir:
  $$
  \mathbf{g} = [g_1, g_2, g_3, g_4, g_5]
  $$
  Burada $g_1$ ilk genin ifade seviyesini, $g_2$ ikinci genin ifade seviyesini temsil eder.

---

## **2. Vektör İşlemleri**

### **2.1 Vektör Toplamı**
İki vektörün toplamı, aynı boyuttaki karşılıklı elemanlarının toplanmasıyla elde edilir:
$$
\mathbf{x} + \mathbf{z} = [x^{(1)} + z^{(1)}, x^{(2)} + z^{(2)}, \dots, x^{(m)} + z^{(m)}]
$$

**Örnek:**
- İki farklı hücrenin gen ifade seviyelerini temsil eden vektörler olsun:
  $$
  \mathbf{g}_1 = [10, 20, 30], \quad \mathbf{g}_2 = [5, 15, 25]
  $$
  Bu iki hücrenin gen ifade seviyelerinin toplamını bulmak için:
  $$
  \mathbf{g}_1 + \mathbf{g}_2 = [10+5, 20+15, 30+25] = [15, 35, 55]
  $$

---

### **2.2 Vektör Farkı**
İki vektörün farkı, aynı boyuttaki karşılıklı elemanlarının çıkarılmasıyla elde edilir:
$$
\mathbf{x} - \mathbf{z} = [x^{(1)} - z^{(1)}, x^{(2)} - z^{(2)}, \dots, x^{(m)} - z^{(m)}]
$$

**Örnek:**
- İki hücrenin gen ifade seviyeleri arasındaki farkı hesaplamak için:
  $$
  \mathbf{g}_1 - \mathbf{g}_2 = [10-5, 20-15, 30-25] = [5, 5, 5]
  $$

---

### **2.3 Skaler Çarpımı**
Bir vektörü bir skalerle çarpmak, vektörün her elemanını o skalerle çarpmak anlamına gelir:
$$
c \cdot \mathbf{x} = [c \cdot x^{(1)}, c \cdot x^{(2)}, \dots, c \cdot x^{(m)}]
$$

**Örnek:**
- Bir hücrenin gen ifade seviyelerini 2 katına çıkarmak için:
  $$
  2 \cdot \mathbf{g}_1 = [2 \cdot 10, 2 \cdot 20, 2 \cdot 30] = [20, 40, 60]
  $$

---

### **2.4 Nokta Çarpımı (Dot Product)**
İki vektörün nokta çarpımı, aynı boyuttaki karşılıklı elemanlarının çarpımlarının toplamıdır:
$$
\mathbf{w} \cdot \mathbf{x} = \sum_{i=1}^{m} w^{(i)} \cdot x^{(i)}
$$

**Örnek:**
- İki farklı hücrenin gen ifade seviyeleri arasındaki benzerliği ölçmek için nokta çarpımı kullanılabilir:
  $$
  \mathbf{g}_1 = [10, 20, 30], \quad \mathbf{g}_2 = [5, 15, 25]
  $$
  Nokta çarpımı:
  $$
  \mathbf{g}_1 \cdot \mathbf{g}_2 = (10 \cdot 5) + (20 \cdot 15) + (30 \cdot 25) = 50 + 300 + 750 = 1100
  $$


### **2.5 Matris ile Vektör Çarpımı**

Bir matrisle bir vektörü çarpmak, vektörün her boyutunu matrisin ilgili satırlarıyla çarpıp toplamak anlamına gelir. Bu işlem sonucunda yeni bir vektör elde edilir.

#### **Örnek:**
Bir proteinin özelliklerini temsil eden bir vektör şu şekilde olsun:
$$
\mathbf{x} = [x^{(1)}, x^{(2)}, x^{(3)}] = [2, 3, 4]
$$

Bu proteinin özelliklerini dönüştürmek için bir dönüşüm matrisi \( W \) kullanalım:
$$
W =
\begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6
\end{bmatrix}
$$

Matris-vektör çarpımını hesaplayalım:
$$
W \cdot \mathbf{x} =
\begin{bmatrix}
1 \cdot 2 + 2 \cdot 3 + 3 \cdot 4 \\
4 \cdot 2 + 5 \cdot 3 + 6 \cdot 4
\end{bmatrix}
=
\begin{bmatrix}
20 \\
47
\end{bmatrix}
$$




Bir **x** vektörünün bir **W** matrisiyle çarpımı, makine öğrenmesi ve derin öğrenme gibi alanlarda sıkça kullanılan bir işlemdir. Özellikle sinir ağlarında (neural networks), verilerin dönüşümü veya ağırlıkların uygulanması için bu işlem kullanılır. Şimdi adım adım açıklayalım:

---

### **1. Matematiksel Gösterim**
Bir **x** vektörü ile bir **W** matrisinin çarpımı şu şekilde tanımlanır:
$$
W \cdot x = y
$$
Burada:
- $ W $: Bir matristir $( m \times n$ boyutunda).
- $ x $: Bir vektördür $( n$-boyutlu sütun vektörü).
- $ y $: Çarpım sonucu elde edilen yeni bir vektördür $( m$-boyutlu).

Matris-vektör çarpımı, **$x$** vektörünün her bir elemanını $W$'nin ilgili satırlarıyla çarparak ve bu çarpımları toplayarak yapılır.

---

### **2. Adım Adım İşlem**
Örnek üzerinden açıklayalım:

#### **Verilenler:**
- $W$: 2x3 boyutunda bir matris.
  $$
  W =
  \begin{bmatrix}
  1 & 2 & 3 \\
  4 & 5 & 6
  \end{bmatrix}
  $$
- $ x$: 3-boyutlu bir sütun vektörü.
  $$
  x =
  \begin{bmatrix}
  2 \\
  3 \\
  4
  \end{bmatrix}
  $$

#### **Çarpım İşlemi:**
Matris-vektör çarpımı şu şekilde hesaplanır:

$$
W \cdot x =
\begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6
\end{bmatrix}
\cdot
\begin{bmatrix}
2 \\
3 \\
4
\end{bmatrix}
=
\begin{bmatrix}
(1 \cdot 2) + (2 \cdot 3) + (3 \cdot 4) \\
(4 \cdot 2) + (5 \cdot 3) + (6 \cdot 4)
\end{bmatrix}
=
\begin{bmatrix}
20 \\
47
\end{bmatrix}
$$

Sonuç olarak, \( y \) vektörü:
\[
y =
\begin{bmatrix}
20 \\
47
\end{bmatrix}
\]

---

### **3. Biyolojik Örnek**
Bu işlemi biyoloji alanında bir örnekle açıklayalım:

#### **Senaryo:**
Bir proteinin özellikleri (örneğin, amino asit kompozisyonu) bir vektörle temsil edilsin:
$$
x =
\begin{bmatrix}
2 \\
3 \\
4
\end{bmatrix}
$$
Bu özellikler, bir sinir ağının gizli katmanına aktarılacak olsun. Gizli katmanın ağırlıkları \( W \) matrisiyle temsil edilsin:
$$
W =
\begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6
\end{bmatrix}
$$

#### **İşlem:**
Protein özelliklerini dönüştürmek için \( W \cdot x \) işlemi yapılır:
$$
W \cdot x =
\begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6
\end{bmatrix}
\cdot
\begin{bmatrix}
2 \\
3 \\
4
\end{bmatrix}
=
\begin{bmatrix}
20 \\
47
\end{bmatrix}
$$

Sonuç olarak, proteinin dönüştürülmüş özellikleri şu şekilde olur:
$$
y =
\begin{bmatrix}
20 \\
47
\end{bmatrix}
$$

Bu sonuçlar, sinir ağının bir sonraki katmanına giriş olarak kullanılabilir.

````python
import numpy as np

# İki hücrenin gen ifade seviyeleri
g1 = np.array([10, 20, 30])
g2 = np.array([5, 15, 25])

# Toplam
toplam = g1 + g2
print("Toplam:", toplam)
````

````python
# Nokta çarpımı
dot_product = np.dot(g1, g2)
print("Nokta Çarpımı:", dot_product)
````

````python
# Dönüşüm matrisi ve vektör
W = np.array([[1, 2, 3], [4, 5, 6]])
x = np.array([2, 3, 4])

# Matris-vektör çarpımı
sonuc = np.dot(W, x)
print("Matris-Vektör Çarpımı:", sonuc)
````
